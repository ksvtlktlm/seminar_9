```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Используем Sentiment Analysis библиотеку для присовоения характеристики отзыву - positive, negative, neutral
```

```python
analyzer = SentimentIntensityAnalyzer()

def vader_sentiment_label(text):
    scores = analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return "positive"
    elif scores['compound'] <= -0.05:
        return "negative"
    else:
        return "neutral"
```

```python
auto_subset = df.sample(50, random_state=10)  # выбор псевдослучайных 100 строк
auto_subset["auto_sentiment"] = auto_subset["review"].apply(vader_sentiment_label)
```

```python
def manual_labeling(subset):
    for index, row in subset.iterrows():
        print(f"Review ({index + 1}/{len(subset)}):\n{row['review']}\n")
        sentiment = input("Enter sentiment (positive/negative/neutral): ")
        subset.at[index, 'manual_sentiment'] = sentiment
    return subset
```
